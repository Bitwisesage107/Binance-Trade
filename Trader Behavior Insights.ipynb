## Feature engineering

### Subtask:
Create relevant features from the raw data that could be useful for modeling.


**Reasoning**:
Extract date and time components from `Timestamp IST` in `historical_df`, calculate absolute values for `Size USD` and `Closed PnL`.



```python
historical_df['year'] = historical_df['Timestamp IST'].dt.year
historical_df['month'] = historical_df['Timestamp IST'].dt.month
historical_df['day'] = historical_df['Timestamp IST'].dt.day
historical_df['hour'] = historical_df['Timestamp IST'].dt.hour
historical_df['minute'] = historical_df['Timestamp IST'].dt.minute
historical_df['day_of_week'] = historical_df['Timestamp IST'].dt.dayofweek

historical_df['abs_size_usd'] = historical_df['Size USD'].abs()
historical_df['abs_closed_pnl'] = historical_df['Closed PnL'].abs()

display(historical_df.head())
```

**Reasoning**:
Create lag features for `Execution Price`, `Size USD`, and `Closed PnL` for each account and coin combination to capture past performance or price movements.



```python
historical_df['price_lag_1'] = historical_df.groupby(['Account', 'Coin'])['Execution Price'].shift(1)
historical_df['size_usd_lag_1'] = historical_df.groupby(['Account', 'Coin'])['Size USD'].shift(1)
historical_df['closed_pnl_lag_1'] = historical_df.groupby(['Account', 'Coin'])['Closed PnL'].shift(1)

display(historical_df.head())
```

## Model training and evaluation

### Subtask:
Train the selected models (Random Forest Regressor and a Gradient Boosting Regressor) to predict trade profitability (Closed PnL) and evaluate their performance using appropriate regression metrics.


**Reasoning**:
Prepare the data for modeling by selecting relevant features, the target variable, and handling missing values. Given the previous EDA showed no missing values in the relevant columns, we can proceed with feature selection and dropping rows with NaNs created by lag features.



```python
features = ['value', 'Size USD', 'price_lag_1', 'size_usd_lag_1', 'closed_pnl_lag_1']
target = 'Closed PnL'

X = merged_df[features]
y = merged_df[target]

# Drop rows with NaN values created by lag features
data = pd.concat([X, y], axis=1).dropna()

X = data[features]
y = data[target]

display(X.head())
display(y.head())
display(X.info())
display(y.info())
```

**Reasoning**:
Split the data into training and testing sets using a time-based split and import the required models and metrics.



```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# For time-based split, we need to use the date information.
# Assuming 'data' dataframe still retains the original index which is sorted by time.
# Find the index to split the data (e.g., 80% for training, 20% for testing)
split_index = int(len(data) * 0.8)

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)
```

**Reasoning**:
Instantiate and train both the RandomForestRegressor and LGBMRegressor models on the training data.



```python
# Instantiate and train RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
print("Random Forest Regressor trained.")

# Instantiate and train LGBMRegressor
lgbm_model = lgb.LGBMRegressor(random_state=42)
lgbm_model.fit(X_train, y_train)
print("LGBM Regressor trained.")
```

**Reasoning**:
Make predictions on the test set using both trained models and evaluate their performance using regression metrics.



```python
# Make predictions
rf_predictions = rf_model.predict(X_test)
lgbm_predictions = lgbm_model.predict(X_test)
print("Predictions made on the test set.")

# Evaluate models
print("\nRandom Forest Regressor Performance:")
print("MAE:", mean_absolute_error(y_test, rf_predictions))
print("MSE:", mean_squared_error(y_test, rf_predictions))
print("R-squared:", r2_score(y_test, rf_predictions))

print("\nLGBM Regressor Performance:")
print("MAE:", mean_absolute_error(y_test, lgbm_predictions))
print("MSE:", mean_squared_error(y_test, lgbm_predictions))
print("R-squared:", r2_score(y_test, lgbm_predictions))
```

```python
import optuna
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Define the objective function for Optuna
def objective(trial):
    # Define hyperparameters to tune
    params = {
        'objective': 'regression',
        'metric': 'mse',
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.5, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
        'lambda_l1': trial.suggest_float('lambda_l1', 1e-8, 10.0, log=True),
        'lambda_l2': trial.suggest_float('lambda_l2', 1e-8, 10.0, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 2, 256),
        'verbose': -1,
        'n_jobs': -1,
        'seed': 42
    }

    # Train the LightGBM model
    model = lgb.LGBMRegressor(**params)
    model.fit(X_train, y_train)

    # Make predictions and evaluate
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)

    return mse

# Create a study object and optimize
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50) # We can adjust the number of trials

# Print the best hyperparameters and the corresponding MSE
print("\nBest hyperparameters:")
print(study.best_params)
print("\nBest MSE:", study.best_value)

# Train the final model with the best hyperparameters
best_lgbm_model = lgb.LGBMRegressor(**study.best_params, random_state=42)
best_lgbm_model.fit(X_train, y_train)

# Evaluate the best model on the test set
best_predictions = best_lgbm_model.predict(X_test)

print("\nBest LightGBM Regressor Performance after tuning:")
print("MAE:", mean_absolute_error(y_test, best_predictions))
print("MSE:", mean_squared_error(y_test, best_predictions))
print("R-squared:", r2_score(y_test, best_predictions))
```
